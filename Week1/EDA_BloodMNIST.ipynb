## 1. Exploratory Data Analysis for BloodMNIST
This notebook performs a full exploratory data analysis (EDA) on the **BloodMNIST** dataset from MedMNIST.

We will inspect dataset structure, visualize images, analyze pixel statistics, and understand class balance.

### Description:
**BloodMNIST** contains **blood cell images** captured from microscopic views, where the goal is to **classify different types of blood cells**.  
This dataset is derived from the **BCCD (Blood Cell Count and Detection) dataset**, which provides labeled images of various human blood cell types.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from medmnist import BloodMNIST

sns.set_theme(style="whitegrid", palette="pastel")
plt.rcParams["figure.figsize"] = (6, 4)
```

## 2. Load Dataset

```python
train_data = BloodMNIST(split='train', download=True)
val_data   = BloodMNIST(split='val', download=True)
test_data  = BloodMNIST(split='test', download=True)

x_train = train_data.imgs
y_train = train_data.labels.flatten()

x_val = val_data.imgs
y_val = val_data.labels.flatten()

x_test = test_data.imgs
y_test = test_data.labels.flatten()

# Combine all labels to check number of unique classes
all_labels = np.concatenate([y_train, y_val, y_test])
num_classes = len(np.unique(all_labels))

print(f"Number of train images: {len(x_train)}")
print(f"Number of val images: {len(x_val)}")
print(f"Number of test images: {len(x_test)}")
print(f"Number of classes: {num_classes}")
print(f"Image size: {x_train[0].shape}")
```

## 3. Convert numeric labels to text

```python
label_map = {0: "Basophil", 1: "Eosinophil", 2: "Erythroblast", 3: "Immature Granulocyte", 4: "Lymphocyte", 5: "Monocyte", 6: "Neutrophil", 7: "Platelet"}
y_train_str = np.array([label_map[y] for y in y_train])
y_val_str   = np.array([label_map[y] for y in y_val])
y_test_str  = np.array([label_map[y] for y in y_test])
```

## 4. Visualize sample images

```python
fig, axes = plt.subplots(3, 5, figsize=(8, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(x_train[i].squeeze(), cmap="gray")
    ax.set_title(f"{y_train_str[i]}", size=10)
    ax.axis("off")
plt.show()
```

## 5. Class distribution

```python
train_df = pd.DataFrame({"label": y_train_str})

plt.figure(figsize=(12, 6))
sns.countplot(x="label", data=train_df, palette="Set2", hue="label", legend=False)
plt.title("Graph of the count of each class on the BloodMNIST")
plt.xlabel("Classes")
plt.ylabel("Number of Images")
plt.show()

plt.figure(figsize=(6, 6))
class_counts = train_df["label"].value_counts()
plt.pie(class_counts, labels=class_counts.index, autopct="%1.1f%%", colors=sns.color_palette("Set2"))
plt.title("Data distribution of BloodMNIST")
plt.show()
```
